```python
import pandas as pd
import numpy as np
from pathlib import Path

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)
```

```python
DATA_DIR = Path("../data/raw")

csv_files = list(DATA_DIR.rglob("*.csv"))

print(f"Found {len(csv_files)} CSV files:\n")

for file in csv_files:
    print(file)
```

```python
def audit_csv(file_path):
    
    print("\n" + "="*80)
    print(f"FILE: {file_path.name}")
    print("="*80)

    try:
        df = pd.read_csv(file_path)

        print(f"\nShape: {df.shape[0]:,} rows x {df.shape[1]} columns")

        print("\nColumns:")
        print(list(df.columns))

        print("\nData Types:")
        print(df.dtypes)

        print("\nMissing Values:")
        missing = df.isnull().sum()
        missing = missing[missing > 0]

        if len(missing) == 0:
            print("No missing values")
        else:
            print(missing.sort_values(ascending=False))

        print("\nSample Rows:")
        display(df.head())

    except Exception as e:
        print("ERROR:", e)
```

```python
for file in csv_files:
    audit_csv(file)
```

```python
summary = []

for file in csv_files:

    try:
        df = pd.read_csv(file)

        summary.append({
            "dataset": file.name,
            "rows": len(df),
            "columns": len(df.columns),
            "missing_cells": int(df.isnull().sum().sum()),
            "memory_mb": round(
                df.memory_usage(deep=True).sum() / 1024 / 1024,
                2
            )
        })

    except:
        pass

summary_df = pd.DataFrame(summary)

summary_df.sort_values(
    "rows",
    ascending=False
)
```

## Key Analysis of Core Tables

```python
# Base directory for Transfermarkt files
base_path = Path("../data/raw/transfermarkt")

# 1. Load the 4 specific datasets
print("⏳ Loading datasets...")
appearances = pd.read_csv(base_path / "appearances.csv")
games = pd.read_csv(base_path / "games.csv")
players = pd.read_csv(base_path / "players.csv")
club_games = pd.read_csv(base_path / "club_games.csv")
print("✅ Datasets loaded successfully.\n")

# ==========================================
# STANDARD INSPECTION (.head() and .info())
# ==========================================
datasets = {
    "appearances": appearances,
    "games": games,
    "players": players,
    "club_games": club_games
}

for name, df in datasets.items():
    print("=" * 80)
    print(f"📋 INSPECTING DATASET: {name}")
    print("=" * 80)
    
    print(f"\n🔹 {name}.head():")
    display(df.head())
    
    print(f"\n🔹 {name}.info():")
    df.info()
    print("\n")

# ==========================================
# SPECIFIC KEY INSPECTION
# ==========================================
print("=" * 80)
print("🔑 SPECIFIC KEY HEAD INSPECTION")
print("=" * 80)

print("\n🔹 appearances[['player_id', 'game_id']].head():")
display(appearances[['player_id', 'game_id']].head())

print("\n🔹 games[['game_id']].head():")
display(games[['game_id']].head())

print("\n🔹 players[['player_id']].head():")
display(players[['player_id']].head())

print("\n🔹 club_games[['game_id', 'club_id']].head():")
display(club_games[['game_id', 'club_id']].head())


# ==========================================
# JOIN VERIFICATION AUDIT
# ==========================================
print("\n" + "=" * 80)
print("🧐 JOIN CLEANLINESS VERIFICATION RESULTS")
print("=" * 80)

# Check 1: appearances.player_id ↕ players.player_id
app_players = set(appearances['player_id'].unique())
tot_players = set(players['player_id'].unique())
matched_players = app_players.intersection(tot_players)

print("\n1. Key Match: appearances.player_id ↕ players.player_id")
print(f"   - Unique players in appearances: {len(app_players):,}")
print(f"   - Unique players in players master: {len(tot_players):,}")
print(f"   - Overlapping/Matching players: {len(matched_players):,}")

if len(app_players) == len(matched_players):
    print("   ✅ CLEAN JOIN: Every single player in appearances exists in the players master file.")
else:
    missing_pct = ((len(app_players) - len(matched_players)) / len(app_players)) * 100
    print(f"   ⚠️  MISMATCH: {len(app_players) - len(matched_players):,} players in appearances ({missing_pct:.2f}%) are MISSING from the players master file.")

# Check 2: appearances.game_id ↕ games.game_id
app_games = set(appearances['game_id'].unique())
tot_games = set(games['game_id'].unique())
matched_games = app_games.intersection(tot_games)

print("\n2. Key Match: appearances.game_id ↕ games.game_id")
print(f"   - Unique games in appearances: {len(app_games):,}")
print(f"   - Unique games in games master: {len(tot_games):,}")
print(f"   - Overlapping/Matching games: {len(matched_games):,}")

if len(app_games) == len(matched_games):
    print("   ✅ CLEAN JOIN: Every single game recorded in appearances exists in the games master file.")
else:
    missing_pct = ((len(app_games) - len(matched_games)) / len(app_games)) * 100
    print(f"   ⚠️  MISMATCH: {len(app_games) - len(matched_games):,} games in appearances ({missing_pct:.2f}%) are MISSING from the games master file.")
```
